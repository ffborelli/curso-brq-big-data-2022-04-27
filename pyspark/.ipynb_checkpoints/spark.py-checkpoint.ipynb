```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("aula-pyspark").getOrCreate()
```

```python
df = spark.read.csv('https://cursos.grandeporte.com.br/files/brq/big-data/datasets/ABT.csv',inferSchema=True,header=True)
```

```python
from pyspark.sql.functions import year
```

```python
col = df.filter(df["Volume"] == 1400400).collect()
```

```python
yeardf = df.withColumn("Year",year(df["Date"]))
```

```python
yeardf.describe().show()
```

```python
yeardf.select("volume","open", "year").show()
```

```python
yeardf.limit(2).show()
```

```python
from pyspark.sql.functions import col, asc, desc
max_df = yeardf.groupBy('Year').max()
max_df.sort( col("Year").desc() ).limit(5).show()
```

```python
from pyspark.sql.functions import month
monthdf = df.withColumn("Month", month("Date"))
monthmax = monthdf.select("Month","Close").groupBy("Month").max()
monthmax.select("Month","max(Close)").orderBy('Month').show()
```

```python
from pyspark.sql.functions import format_number
#df.select("open", format_number( df["volume"].cast('float'),2).alias("open_float") ).show()

dfFormat = df.withColumn( "open_float", format_number( df["open"].cast('float'),2) )

dfFormat.select("open","open_float").limit(2).show()
```
